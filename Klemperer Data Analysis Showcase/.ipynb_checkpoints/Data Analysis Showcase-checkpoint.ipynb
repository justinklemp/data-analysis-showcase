## TED X Data (Part 2) - Groupby & Aggregation

Load up 'ted_main.csv' data by using a pd.read_csv() function. Title this as ted_df

```python
ted_df = pd.read_csv('ted_main.csv')
ted_df.head()
```

We will now use the TED X Data and perform a groupby/aggregation function to assess the most viewed and commented TED X talks, also showing us the speaker and the title of the video.

"At a high level, when we have repeated values within a DataFrame, we can group these together and compute aggregate statistics on each of the repeated values. This is an important step for summarizing large-scale data and reducing it down to a size and shape where we can analyze and visualize."(KEEGAN2018) 

To start, create an object in groupby, passing a list of 'main_speaker' and 'title' usings the .groupby() function. Label this object tedx_groupby

```python

tedx_groupby = ted_df.groupby(['main_speaker','title'])
```

Next, you have to define the aggregation dictionary. For this groupby, we want to use the 'views' and 'comments' inside of the dictionary, and use the :.np.sum function to allow us to see a total count of views and comments for all TED X videos

```python

ted_agg = {'views':np.sum,'comments':np.sum}
```

Finally, combine the aggregation dictionary with the groupby object by creating an object, then using the groupby.agg() function. Then, to find the top viewed videos, use .sort_values() with 'views' using an ascending function. Show the top 15 videos

```python

ted_top_views = tedx_groupby.agg(ted_agg)
                                            
ted_top_views.sort_values('views', ascending = True).tail(15)
```

To double check if this is correct, we will use a .max() function on the dataset to confirm that 47227110 is the correct top viewed video

```python
ted_df['views'].max()
```

As we can see from our groupby aggregation, Ken Robinson's "Do schools kill creativity?" is the highest viewed video

Next, we will run a similar process to allow us to see what the most commeneted video is out of the entire database. We can use the same process to find this as well.

To start, create an object in groupby, passing a list of 'main_speaker' and 'title' usings the .groupby() function. Label this object tedx_groupby2

Then, define the aggregation dictionary. For this groupby, we want to use the 'views' and 'comments' inside of the dictionary, and use the :.np.sum function to allow us to see a total count of views and comments for all TED X videos. Label this as tedd_agg2

Also combine the aggregation dictionary with the groupby object by creating an object, then using the groupby.agg() function. Label this as ted_top_comments

```python
tedx_groupby2 = ted_df.groupby(['main_speaker','title'])

ted_agg2 = {'views':np.sum,'comments':np.sum}

ted_top_comments = tedx_groupby2.agg(ted_agg2)
```

Lastly, to find the top commented videos, use .sort_values() with 'views' using an ascending function. Show the top 15 videos.

```python
                                  
ted_top_comments.sort_values('comments', ascending = True).tail(15)
```

To double check if this is correct, we will use a .max() function on the dataset to confirm that 6404 is the correct top commented video

```python
ted_df['comments'].max()
```

As we can see from the data through our groupby/aggregation, Richard Dawkins, "Militant Atheism" is the most commented video.

## NHL Stats (Part 3) - Joinging/Merging + Dropping Missing Data

Load up the NHL data by creating a callable object to pd.read_csv() both datasets 'Scoring.csv' and 'Master.csv' - First, use pd.read_csv() with 'Scoring.csv' and label it scoring_df

```python
scoring_df = pd.read_csv('Scoring.csv')
scoring_df.head(5)
```

We can fill the 'NaN' values with 0s by using the .fillna() method and passing a value inside in order to replace all the 'NaN' values with another value (0). We will fill any rows of data that include a player who does not have any post season goals or assists by passing the 'PostPts' vlaue inside to replace all 'NaN'.

```python
clean_scoring_df = scoring_df['PostPts'].fillna(0)
```

Now, use pd.read_csv() with 'Master.csv' and label it master_df

```python
master_df = pd.read_csv('Master.csv')
master_df.head(5)
```

Use the pd.merge() function on both 'scoring_df' and 'master_df'.

"pd.merge can be used by passing it five arguments:

left - the left-hand DataFrame
right - the right-hand DataFrame
left_on - the name of the column containing the values to join on in the left DataFrame
right_on - the name of the column containing the values to join on in the right DataFrame
how - options include "left", "right", "inner" and "outer" - (KEEGAN2018)

Use 'playerID' for both left_on and right_on and perform a 'left' merge to contain most of all the data in our left DataFrame, even though it may reduce some of the data on our master_df dataset.

```python
values_land_left_df = pd.merge(scoring_df,master_df,left_on='playerID',right_on='playerID',how='left')

values_land_left_df.tail()
```

We can use the .dropna(subset = [ ]) to allow us to only drop rows of data that are missing a specific colum of data. For this, let's look at 'deathYear' to remove any rows of data that do not show which year they passed away. This will give us an indication for how old some of the first NHL players were back in the day.

```python
droped = values_land_left_df.dropna(subset=['deathYear'])
droped.head()
```
